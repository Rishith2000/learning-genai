```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv(override=True)
client = Anthropic()

resp = client.messages.create(
    model="claude-haiku-4-5",
    max_tokens=256,
    messages=[{"role": "user", "content": "In one sentence, what is RAG?"}],
)
```

```python
messages = []   # persists across cells — this is your memory

def chat(user_input):
    messages.append({"role": "user", "content": user_input})
    resp = client.messages.create(
        model="claude-haiku-4-5",
        max_tokens=512,
        system="You are a helpful, concise assistant.",
        messages=messages,
    )
    reply = resp.content[0].text #"".join(b.text for b in resp.content if b.type == "text")
    messages.append({"role": "assistant", "content": reply})
    print(f"claude> {reply}")
    print(f"[turns: {len(messages)} | tokens in: {resp.usage.input_tokens}]")
```

```python
chat("What is RAG?")
```

```python
chat("my name is rishith reddy")
```

```python
chat("What is my name ?")
```
