```python
from sentence_transformers import SentenceTransformer
import numpy as np


model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

```python
#Knowledge base documents
docs = [
    "The capital of France is Paris.",
    "Python is a popular programming language for data science.",
    "The Eiffel Tower is located in Paris, France.",
    "Photosynthesis is how plants convert sunlight into energy.",
    "Machine learning models learn patterns from data.",
]
```

```python
doc_embeddings = model.encode(docs, normalize_embeddings=True)
print("doc_embeddings shape:", doc_embeddings.shape)   # (5, 384)
```

```python
query = "Where is the Paris?"
query_embedding = model.encode(query, normalize_embeddings=True)   # (384,)
```

```python
scores = doc_embeddings @ query_embedding    # (5,)
print("scores:", np.round(scores, 3))
```

```python
ranked = np.argsort(scores)[::-1]     # indices, best first
for rank, i in enumerate(ranked, 1):
    print(f"{rank}. ({scores[i]:.3f})  {docs[i]}")
```

```python
a=np.argmax(scores)
print(a)
```
